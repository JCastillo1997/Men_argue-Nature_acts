```python
df_pop = df_pop[~df_pop['Macrotrends Data Download'].str.contains(r'[A-Za-z]')]
```

```python
# Split the column into two new columns
df_split = df_pop['Macrotrends Data Download'].str.split(',', expand=True)

# Assign the new columns to the DataFrame
df_pop['Date'] = df_split[0]  # Extract the date
df_pop['Data'] = df_split[1]  # Extract the data after the first comma

# Convert 'Date' to datetime format
df_pop['Date'] = pd.to_datetime(df_pop['Date'], format='%Y-%m-%d')

# Display the resulting DataFrame with the extracted columns
print(df_pop[['Date', 'Data']])
```

```python
# Convert 'Date' to year, month, and day columns
df_pop['Year'] = df_pop['Date'].dt.year
df_pop['Month'] = df_pop['Date'].dt.month
df_pop['Day'] = df_pop['Date'].dt.day

# Drop the original 'Date' column if needed
# df_pop.drop(columns=['Date'], inplace=True)
```
